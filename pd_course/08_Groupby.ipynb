## Groupby

```python
# lets redo the above finding sales for ps3 & x360

sales.groupby('Platform').sum()
```

```python
# whats the average for each platform across the regions?

round(sales.groupby('Platform').mean()*10**7, 2)
```

```python
# Now what if we wanted to call all playstaions playstatino and all xboxes xbox and get the sum?

sales.Platform.unique()
```

```python
platform_names = {
    'X360' : 'XBox',
    'XOne' : 'XBox',
    'PS3' : 'Playstation',
    'PS4' : 'Playstation'
}

platform_names
```

```python
sales.set_index('Platform').groupby(platform_names).sum()
```

```python
# **** to do the above operation, MUST set the index to the col that contains all the names first *******

# in this example we need to set index to Platform, then groupby platform names
# note this allows the sales df to remain UNCHANGED
sales
```

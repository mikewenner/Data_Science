## Fine Tuning Aggregates / Multi Aggs

```python
# summarize the sum avg and stdev of sales as well as the number of games published by each publisher within each genre

studios.head()
```

```python
studios.groupby(['Genre', 'Publisher']).agg(['count', 'mean', 'std', 'sum'])
```

```python
# if we want to sort we need to sort by the multi index
# so to sort by sum:

studios.groupby(['Genre', 'Publisher']).agg(['count', 'mean', 'std', 'sum']).sort_values(by=('Global_Sales', 'sum'), ascending=False)
```

```python
# or can call the describe on the groupby
studios.groupby(['Genre', 'Publisher']).describe()
```

## Naming Aggs

### here we can change the column name and the function it calls

```python
studios.groupby(['Genre', 'Publisher']).agg(
    tot_revenue = ('Global_Sales', 'sum'),
    num_games = ('Global_Sales', 'count'),
    stdev_revenue = ('Global_Sales', 'std'),
    average_revenue = ('Global_Sales', 'mean'),
).sort_values(by=('tot_revenue'), ascending=False)

# note the Multi level column disappears, so when sort values only need to specify tot_revenue
```

```python
# find all the games whose publisher has sold more than 50M in north american sales within the games genre
games.groupby(['Publisher', 'Genre']).filter(lambda sg: sg['NA_Sales'].sum() > 50)
```

```python
#without the lambda function

def more_than_50(df):
    return df['NA_Sales'].sum() > 50
```

```python
games.groupby(['Publisher', 'Genre']).filter(more_than_50)
```

## Transform

### in group aggreations

```python
# the transform aligns the sums with the original df so we can filter

sales_sum = games.groupby(['Publisher', 'Genre'])['NA_Sales'].transform('sum')
```

```python
filtered_games = games[sales_sum > 50]
filtered_games
```

## Summary

### .agg() --> will reduce the dimensionality of the df

### .filter() --> include/exclude w/o changint the dims

### .transform() --> changes the values w/o changing shape

## Apply method
* used in conjunction with groupby, can apply a function 
* applied to the SUBGROUP

```python
# EX, start by slicing out subset of data from games, just want the PS3

ps3_data = games.loc[games.Platform == 'PS3', ['Name', 'Genre', 'EU_Sales', 'Global_Sales']]
ps3_data
```

```python
# if we wanted to get info on the genre sub groups:

# ps3_data.groupby('Genre').apply(lambda x: print(x.info(), '\n'))
```

```python
# group by the genre and apply function

ps3_data.groupby('Genre').apply(lambda sg: 'solid' if sg.EU_Sales.sum() > 50 else 'weak')
```

```python
# so this look at the genres as sub groups and applies the sg function
```

```python
# we can also break this down further by creating a function for variance
# variance = std/mean

def sales_detail(sg):
    level = 'solid' if sg.EU_Sales.sum() > 50 else 'weak' # did this above
    var = 'volatie' if sg.EU_Sales.std() / sg.EU_Sales.mean() > 2 else 'steady'
    return (var, level + ' sales')
    
```

```python
# put into the apply

ps3_data.groupby('Genre').apply(sales_detail)
```

# Challenge exercise

### Find the players in our dataset that meet these criteria

1.   they're English (*nationality*), and
2.   their market value is more than twice the average market value in the league (*market_value*), and
3. they either have more than 4,000 views (*page_views*) or are a new signing (*new_signing*) but not both

```python
players.head(3)
```

```python
# Assign variables to each condition above

english_nationality = players.nationality =='England'
english_nationality.value_counts()
```

```python
market_val = players.market_value > (players.market_value.mean() *2)
market_val.value_counts()
```

```python
# Third condition --> XOR (^)
third_cond = (players.page_views > 4000) ^ (players.new_signing == 1)
third_cond.value_counts()
```

```python
# Putting it all together
# All conditions combine with AND

players.loc[english_nationality & market_val & third_cond]
```

## Transform

### applies a function to a df w/o changing size or shape of the df

```python
# ex using transform

# what if we wanted to vary the characters (types) in the whole df?

# we can randomly choose different string types

# import the choice function from random

# define a function that chooses

# call the function on the df
```

```python
from random import choice
```

```python
def random_case(x):
    functions = [x.str.swapcase, x.str.lower, x.str.upper, x.str.title]
    
    return choice(functions)()
```

```python
players.select_dtypes(include=object).transform(random_case).head()
```

```python
# run multiple test cases

for i in range(5):
    print(players.select_dtypes(include=object).transform(random_case).head())
```

```python
# 1. create a functions that defines popularity categories.

def get_popularity(x):
    if x < 220:
        return 'unknown'
    elif x < 600:
        return 'kind of popular'
    elif x < 2000:
        return 'popular'
    else:
        return 'super-popular'
```

```python
# test the function
get_popularity(1000)
```

```python
# 2. apply the function to page views

players.page_views.apply(get_popularity)
```

```python
# 3. add new col to df with the results, call it popularity

players['popularity'] = players.page_views.apply(get_popularity)
players.head()
```

```python
# 4. How many super popular players are there?

players.popularity.value_counts()
```

```python
# if want to see all the players
players[players.popularity == 'super-popular']
```

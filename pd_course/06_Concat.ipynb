```python
# we can see how index starts over when eng schools start
# we can use the keys parm to distinguish

new_df = pd.concat([ivies, eng], keys = ['ivies', 'engineering'])
new_df
```

# Merge - like sql 

* pd.merge defaults to inner join --> only where both occur
    - pd.merge(how='inner')
* left_on='col_name', right_on='col_name'

```python
# read in the regions csv

regions_path = '/Users/mikewenner/Desktop/CSV/regions.csv'
regions = pd.read_csv(regions_path)
regions
```

```python
# look at schools of liberal & state we made earlier
lib_state.head()
```

```python
# merge them together
pd.merge(lib_state, regions)
```

```python
# note this occurs with the school name as the merge column

# its the same as 

pd.merge(lib_state, regions, on="School Name")
```

```python
# If the column names we want to merge on are different ... 
# the on parameter is now different

mid_path = '/Users/mikewenner/Desktop/CSV/mid_career_salaries.csv'
mid = pd.read_csv(mid_path)
mid
```

```python
# note that school name is lower case with under score

pd.merge(lib_state, mid, left_on="School Name", right_on="school_name")
```

```python
# pd.concat([lib_state, mid], axis=1) --> will error because the column names are different
```

```python
# drop the school_name column

pd.merge(lib_state, mid, left_on="School Name", right_on="school_name").drop(columns=['school_name'])
```

## Multiple Conditions

```python
# Find all players the are LB's
players.position == 'LB'
```

```python
players[players.position == 'LB']
```

```python
# If want all LB's that are less than 25
players[(players.position == 'LB') & (players.age <= 25)]

# MUST wrap each condition in () or it will error out
```

```python
# 3 conditions: LB's 25 or younger and market value 10+
players[
    (players.position == 'LB') &
    (players.age <= 25) &
    (players.market_value >= 10)
    
]
```

```python
# 4 conditions: LB's 25 or younger and market value 10+, and NOT from Aresenal ot Tottenham
players[
    (players.position == 'LB') &
    (players.age <= 25) &
    (players.market_value >= 10) &
    ~(players.club.isin(['Tottenham', 'Arsenal']))
    
]

# **** NOTE THE ~ NEGATES AN ISIN *****
```

```python
# find Arsenal right backs and Chelsea goal keepers

arsenal_player = players.club == 'Arsenal'
arsenal_player
```

```python
right_back = players.position == 'RB'
right_back
```

```python
chelsea_and_GK = (players.club == 'Chelsea') & (players.position == 'GK')
chelsea_and_GK
```

```python
chelsea_and_GK.value_counts()
```

```python
# Combine using the or operator --> pipe
players.loc[arsenal_player & right_back | chelsea_and_GK]
```

# Challenge exercise

### Find the players in our dataset that meet these criteria

1.   they're English (*nationality*), and
2.   their market value is more than twice the average market value in the league (*market_value*), and
3. they either have more than 4,000 views (*page_views*) or are a new signing (*new_signing*) but not both

```python
players.head(3)
```

```python
# Assign variables to each condition above

english_nationality = players.nationality =='England'
english_nationality.value_counts()
```

```python
market_val = players.market_value > (players.market_value.mean() *2)
market_val.value_counts()
```

```python
# Third condition --> XOR (^)
third_cond = (players.page_views > 4000) ^ (players.new_signing == 1)
third_cond.value_counts()
```

```python
# Putting it all together
# All conditions combine with AND

players.loc[english_nationality & market_val & third_cond]
```

## 2D Indexing

```python
# Looks at chelsea players 23 yr old & younger
chelsea_23under = (players.club == 'Chelsea') & (players.age.le(23))
chelsea_23under.value_counts()
```

```python
players[chelsea_23under]
```

```python
# If we wanted to see only certian columns in the dataset
# use .loc and then pass list of columns as 2nd arg

players.loc[chelsea_23under, ['position', 'market_value']]
```

```python
# what if we wanted to see all columns the start with 'p'
# can use the startswith method

p_cols = players.columns.str.startswith('p')
p_cols
```

```python
# Now we have another mask we can pass

players.loc[chelsea_23under, p_cols]
```

```python
## Lookup Method

players.loc[[0, 132], ('name', 'market_value')]
```

```python
players.lookup([0, 132], ['name', 'market_value'])
```

## Sort vs reorder

```python
# reindex() method

players_lite = players.iloc[:4, :4]
players_lite
```

```python
# what if we had a requirement to see the df in the following way:

# row order: 2, 1, 3, 0
# columns order: age, name, position, club
```

```python
#pass python lists into the reindex method
players_lite.reindex(index=[2, 1, 3, 0], columns=['age', 'name', 'position', 'club'])
```

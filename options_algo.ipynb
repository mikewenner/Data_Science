## Filter through puts to sell that offer sufficient premiu as compared to stock price
* puts that are bid
* option premium that is at least 1% of the underlying stock price
* OTM puts at least $5 OTM or greater
* Only options with 25-45 DTE

```python
# Search parms
# here we select a symbol and an expiration, we can omit the expy if we want to filter through all available contracts
symbol = 'MSFT'
expy = '2023-08-11'
```

```python
# Geting the chain data
chain_df = openbb.stocks.options.chains(symbol=symbol) #here we could add expy=expy parm for a single expiration
chain_df.head()
```

```python
# Create a new df to filter out and look at just the puts
puts_df = chain_df[chain_df.optionType == 'put']
puts_df
```

```python
# Now we need the grab the price of the underlying
# use the tob to get active stock price data
tob_price = openbb.stocks.tob(symbol)
tob_price
```

```python
# Above tob_price is a tuple that gives both size and bid & ask price data
# make df from the tob & assign bid & ask cols
bid,ask = tob_price
quote_df = bid.join(ask, lsuffix= ': Bid', rsuffix = ': Ask')
quote_df
```

```python
# Calc spot price by averaging the bid and ask together
spot = round((quote_df.iloc[0][1] + quote_df.iloc[0][3]) / 2,2)
spot
```

```python
# Add the calced spot price to puts df for ref

puts_df['spot'] = spot
puts_df
```

```python
# Filter for OTM only puts (strike is less than spot)

otm_puts = puts_df[puts_df['strike'] < puts_df['spot']]
otm_puts
```

```python
# Remove puts with no bid
otm_puts = otm_puts[otm_puts.bid >0.01]
otm_puts
```

```python
# add in mid price of the puts by averageing the bid/ask of the put
otm_puts['mid_price'] = round(otm_puts.bid + otm_puts.ask / 2, 2)
otm_puts
```

```python
# Find where the put price is greater than 1% of the spot
otm_puts = otm_puts[otm_puts.mid_price / otm_puts.spot > 0.01]
otm_puts
```

```python
# add in OTM col to show how far out of the money the put is
otm_puts['otm'] = otm_puts.spot - otm_puts.strike
#otm_puts.drop(columns=['diff'], inplace=True)
otm_puts
```

```python
# Filter for at least $5 OTM
otm_puts = otm_puts[otm_puts.otm > 5]
otm_puts
```

```python
# So now we have a df of all the puts that are greater than $5 OTM and have a mid price (premium) that is at least 1% of the underlying spot

# But we also see that there are contracts wat out into 2025 in the df

# Now we want to filter the time down to something closer to today --> 25-45 DTE
```

```python
# Add time filtering to expiration
otm_puts.info()
```

```python
# right now, expiration is an object
# need convert to datetime that we can opeerate on
otm_puts.expiration = pd.to_datetime(otm_puts.expiration)
```

```python
otm_puts.info()
```

```python
# add a dte col by subtracting the expy from today
otm_puts['dte'] = otm_puts.expiration - pd.datetime.now()
otm_puts
```

```python
# now convert the time delta (dte col) to just the days 
otm_puts.dte = otm_puts.dte.dt.days
otm_puts
```

```python
# Grabbing the puts between 25-45 dte
otm_puts_25_45 = otm_puts[otm_puts.dte.between(25,45)]
otm_puts_25_45
```

```python
# so now we have a list of options that are at least $5 OTM, 
# have a premium of 1% or more as compared to spot of stock 
# and are within 25-45 dte 
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import shap
import matplotlib.pyplot as plt


```

```python
# Generate a synthetic dataset
np.random.seed(42)
X = np.random.rand(100, 3)
y = X[:, 0] * 2 + X[:, 1] * -1 + X[:, 2] * 3 + np.random.randn(100) * 0.5

# Convert to pandas DataFrame for easier handling
X_df = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Initialize SHAP explainer and compute SHAP values for the test set
explainer = shap.Explainer(model, X_train)
shap_values = explainer(X_test)

# Plot the SHAP values with a beeswarm plot
shap.summary_plot(shap_values, X_test, plot_type="beeswarm")

# Show the plot
plt.show()
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")
```

```python
plt.figure(figsize=(10, 8), dpi=100)
shap.summary_plot(shap_values, X_test, plot_type="beeswarm")
```

```python


# Flatten the SHAP values array and plot a histogram
shap_values_flattened = shap_values.values.flatten()
plt.hist(shap_values_flattened, bins=30)
plt.title("Distribution of SHAP Values")
plt.show()
```

```python
shap.summary_plot(shap_values, X_test, plot_type="beeswarm", plot_size=(10, 8), max_display=10)
```

```python


# Flatten the SHAP values and get corresponding feature names
shap_values_flattened = shap_values.values.flatten()
feature_names = np.repeat(shap_values.feature_names, len(X_test))

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(shap_values_flattened, feature_names, alpha=0.5)
plt.axvline(x=0, color='grey')
plt.xlabel('SHAP Value')
plt.ylabel('Feature')
plt.title('Custom Beeswarm Plot')
plt.show()
```

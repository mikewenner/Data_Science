```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from openbb_terminal.sdk import openbb
from scipy.stats import poisson, expon
```

# Determine the average arrival rate of calls per hour using an estimate.

```python
# Parms for simulation
call_parm = 50 # Estimate of averge calls per hour
avg_time_call = 10 # Average handling time, in minutes per call
work_hours = 8 # number of hours per shift worked in a day
shift_duration = work_hours * 60 # Duration of the shift in minutes
```

```python
shift_duration
```

# Use the Poisson distribution to generate 1000 random numbers based on the provided mean
## These random numbers represent the number of calls

```python
#poisson.rvs->this is the random variable sampling function specifically for the poisson distribution
#mu->the mean parameter of the distribution (average calls for the shift)
#size->size of the output,meaning rvs should generate 1000 random values from the poisson distribtion
num_calls = poisson.rvs(mu=call_parm * shift_duration / 60, size=1000)
#num_calls
```

# Now lets find the mean number of calls generated from the above poisson distribution 

```python
#np.ceil-> rounds the value to the next highest number to ensure an integer
#np.mean-> takes the mean value
mean_num_calls = np.ceil(np.mean(num_calls)).astype(int)
mean_num_calls
```

# Simulate call handling times 
## Assume an exponential function/distrubution for the call handle time

```python
#expon.rvs-> uses the expon module from scipy and the random variable sampling
    # ***The exponential distribution is commonly used to model time between events that occur at a constant rate
#scale-> represents the avg time between events, here calls
#size-> represents the size of the output, here we generate random variables equal to the mean_num_calls variable
call_time = expon.rvs(scale=avg_time_call, size=mean_num_calls)
```

```python
# find the total handling time for the calls in the sim
total_call_time = np.sum(call_time)
total_call_time
```

```python
# Now find the number of people needed for the shift
required_workers = np.ceil(total_call_time / shift_duration).astype(int)
required_workers
```

```python
plt.hist(num_calls, bins='auto', edgecolor = "black")
plt.xlabel('Number of Calls')
plt.ylabel('Frequency')
plt.title('Number of Calls - Poisson Distribution')
plt.show()
```

```python
plt.hist(call_time, bins='auto', edgecolor='black')
plt.xlabel('Call Time')
plt.ylabel('Frequency')
plt.title('Call Time - Exponential est')
plt.show()
```
